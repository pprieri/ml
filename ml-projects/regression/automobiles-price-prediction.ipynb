## Pre-processing with Pipelines

Similar to how a pipeline bundles together preprocessing and modeling steps, we use the ColumnTransformer class to bundle together different preprocessing steps. The code below:

imputes missing values in numerical data, and
imputes missing values and applies a one-hot encoding to categorical data.

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler 

# Preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer_num',SimpleImputer(strategy='constant')),
    ('std_scaler',StandardScaler())
])

# Preprocessing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('one_hot', OneHotEncoder(handle_unknown='ignore'))
])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

### Define the Model

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=0,n_jobs=-1)
```

### Create and Evaluate the Pipeline

Finally, we use the Pipeline class to define a pipeline that bundles the preprocessing and modeling steps. There are a few important things to notice:

With the pipeline, we preprocess the training data and fit the model in a single line of code. (In contrast, without a pipeline, we have to do imputation, one-hot encoding, and model training in separate steps. This becomes especially messy if we have to deal with both numerical and categorical variables!)
With the pipeline, we supply the unprocessed features in X_valid to the predict() command, and the pipeline automatically preprocesses the features before generating predictions. (However, without a pipeline, we have to remember to preprocess the validation data before making predictions.)

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                  ('model', model)
                                 ])
```

```python
#Helper function to calculate the model's performance

def r2_score_pipeline(model, X, y):
# Bundle preprocessing and modeling code in a pipeline
    my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                  ('model', model)
                                 ])

    # Preprocessing of training data, fit model 
    # + Cross-validate the model
    train_score = cross_val_score(my_pipeline, X_train, y_train, cv=5)


    # We now fit the pipeline for one last model evaluation on Test set
    my_pipeline.fit(X_train,y_train)

    test_preds = my_pipeline.predict(X_test)
  
    return np.mean(train_score), test_preds
```

## Principal Components Analysis

```python
from sklearn.decomposition import PCA

features = ["highway-mpg", "engine-size", "horsepower", "curb-weight"]

# X_train_red = X_train.loc[:, features]
# X_test_red = X_test.loc[:, features]

# # Standardize
# std=StandardScaler()
# X_train_scaled = std.fit_transform(X_train_red)
# X_test_scaled = std.transform(X_test_red)

X = df.loc[:,features]

std=StandardScaler()
X_scaled = std.fit_transform(X)

# Create principal components
pca = PCA()

X_pca = pca.fit_transform(X_scaled)

# Convert to dataframe
component_names = [f"PC{i+1}" for i in range(X_pca.shape[1])]
X_pca = pd.DataFrame(X_pca, columns=component_names)

X_pca.head()
```

After fitting, the PCA instance contains the loadings in its components_ attribute

```python
loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=component_names,  # so the columns are the principal components
    index=X.columns,  # and the rows are the original features
)
loadings
```

```python
loadings.to_csv('PC_loadings.csv')
```

The first component (PC1) shows a contrast between large, powerful vehicles with poor gas milage, and smaller, more economical vehicles with good gas milage. We might call this the "Luxury/Economy" axis. The next figure shows that our four chosen features mostly vary along the Luxury/Economy axis.

```python
def plot_variance(pca, width=8, dpi=100):
    # Create figure
    fig, axs = plt.subplots(1, 2,figsize=(7,4))
    n = pca.n_components_
    grid = np.arange(1, n + 1)
    # Explained variance
    evr = pca.explained_variance_ratio_
    axs[0].bar(grid, evr)
    axs[0].set(
        xlabel="Component", title="% Explained Variance", ylim=(0.0, 1.0)
    )
    # Cumulative Variance
    cv = np.cumsum(evr)
    axs[1].plot(np.r_[0, grid], np.r_[0, cv], "o-")
    axs[1].set(
        xlabel="Component", title="% Cumulative Variance", ylim=(0.0, 1.0)
    )
    # Set up figure
    fig.set(figwidth=8, dpi=100)
    return axs
```

```python
def pca_inspired(df):
    X = pd.DataFrame()
    X["PCA_inspired_1"] =  (df["horsepower"] / df["curb-weight"] ) 
    return X
```

## k-Means Clustering
Unsupervised algorithm to create features: k-means clustering. These features can sometimes be effective at untangling complicated spatial relationships.

```python
from sklearn.cluster import KMeans
features = ["highway-mpg", "engine-size", "horsepower", "curb-weight"]

def cluster_labels(df, features, n_clusters=2):
    X = df.copy()
    X_scaled = X.loc[:, features]
    X_scaled = (X_scaled - X_scaled.mean(axis=0)) / X_scaled.std(axis=0)
    kmeans = KMeans(n_clusters=n_clusters, n_init=50, random_state=0)
    X_new = pd.DataFrame()
    X_new["Cluster"] = kmeans.fit_predict(X_scaled)
    return X_new
```

```python
def create_features(X_train, X_test):
    
    X = pd.concat([X_train,X_test])
    
    X = X.join(pca_inspired(X))
    X = X.join(cluster_labels(X,features))
    
    #Reform splits
    X_test = X.loc[X_test.index, :]
    X.drop(X_test.index, inplace=True)
    
    return X, X_test

X_train, X_test = create_features(X_train, X_test)
```

## Averaging of Ridge Linear Regression, NN, kNN and XGBoost
Usually, a stacking/combination of diverse predictors of good models will yield better results, the assumption being: good and diverse models tend to agree on good predictions and disagree on wrong ones (thus compensating the effects between them)

```python
models = [xgb, neigh, mlp, Ridge()]
stacked_preds =  np.zeros([y_test.shape[0],len(models)])

train_preds = np.zeros([y_train.shape[0],len(models)])

for count,model in enumerate(models):
    
    my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)
                             ])

    # We now fit the pipeline for one last model evaluation on Test set
    my_pipeline.fit(X_train,y_train)
    
    #We get test predictions from model 
    preds = my_pipeline.predict(X_test)
    
#     print('R2 score of {} is {}'.format(model,r2_score(y_test,preds)))
    #array with test data points, where each column is a pred by a different model
    stacked_preds[:,count]=preds
    
    #we save also train preds for future step
    preds_train = my_pipeline.predict(X_train)
    train_preds[:,count] = preds_train

stacked_preds.shape    
```

```python
#We calculate average of test predictions
avg_preds = np.mean(stacked_preds,axis=1)

print('R2 score of stacked models {}'.format(r2_score(y_test, avg_preds)))
```

#### => by performing a simple averaging of models, we were able to boost the R2 score from 0.76 to 0.78

### Meta model Linear Regression based on predictions of previous models

```python
ridge_meta_model = RandomForestRegressor(random_state=0,n_jobs=-1,n_estimators=2500)
# ridge_meta_model = Ridge()

# We now fit the pipeline for one last model evaluation on Test set
ridge_meta_model.fit(train_preds,y_train)

#We get final predictions from test for Meta Ridge model
preds = ridge_meta_model.predict(stacked_preds)

print('R2 score of Meta Ridge model {}'.format(r2_score(y_test, preds)))
```

#### => The Level 2 Ridge Meta model increased even more the performance (from 0.7 to 0.8).
